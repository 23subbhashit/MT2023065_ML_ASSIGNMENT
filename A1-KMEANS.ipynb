```python
col = data.columns[data.isna().any()].tolist()
col = col + ["sofifa_id","short_name" ]
```

```python
x = data.drop(col,axis=1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical_features = ["preferred_foot","work_rate","body_type" ]
for each in categorical_features:
    x[each] = encoder.fit_transform(x[each])
```

```python
def plot_clusters(X, centers, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=10, label='Data Points')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Centroids')
    plt.title(title)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
X= x.iloc[:, :].values
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)     
plt.plot(range(1,11), wcss,)
plt.title('The Elbow Method')
plt.xlabel('No of clusters')
plt.ylabel('Intertia for that cluster')
plt.show() 
```

```python
k = 3
kmeans = KMeans(n_clusters = k, random_state = 0).fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
plot_clusters(X, kmeans.cluster_centers_, kmeans.labels_, "K-means optimal clustering")
```
